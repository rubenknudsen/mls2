```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline  
```

```python

n=1000
x = np.linspace(0,40, num=n)-20
y = 400 - np.power(x,2)
#print(y)
y += np.random.normal(scale=100, size=n)
y[0:-1:2] = np.random.randint(-200,600,(int(n/2)))

plt.scatter(x,y)
```

```python
extent = (x.min(), x.max(), y.min(), y.max())

plt.hexbin(x,y, gridsize=10, bins=3,  extent=extent)
```
